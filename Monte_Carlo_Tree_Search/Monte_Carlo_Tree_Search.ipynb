```python
def chat_completion_request(prompt):

  messages = [
            {
                "role": "user",
                "content": [{"type": "text", "text": prompt}]
            }
        ]
  response =  completion(
    model=MODEL_NAME,
    temperature=1.0,
    max_tokens=1500,
    messages=messages
    )
  return response.choices[0].message.content.strip()

```

```python
def get_critique(question,draft_answer):
  prompt=(
      f"Question:{question}\n"
      f"Draft Answer:{draft_answer}\n"
      "Please critique the draft answer ."
      "Do a careful assesment of whether the answer is correct or not. ,why? "
      "consider multiple ways of verifying the correctness of the answer"
      "Do point out every flaw and hold the draft answer to the highest standard"
      "Do provide specific recommendation to improve the answer."
      "Do think step by step"
      "Do not provide a revised answer"
      "Do not provide a revised answer"
      "Do not provide a revised answer"
  )
  return chat_completion_request(prompt)
```

```python
def improve_answer(question,draft_answer,critique):
  prompt=(
      f"Question:{question}\n"
      f"Draft Answer:{draft_answer}\n"
      f"Critique:{critique}\n\n"
      "Please improve the draft answer based on this critique. Follow this format"
      "Reasoning Process : <step-by-step reasoning process>\n"
      "Verification: <verification of the facts>\n"
      "Final Answer: <the improved and verified answer>\n "
  )
  improved_response=chat_completion_request(prompt)
  return improved_response
```

```python
import re
def rate_answer(question,answer):
  prompt=(
      f"Question: {question}\n"
      f"Answer: {answer}\n\n"
      "As an expert on this topic, please provide a detailed critique of the answer ,pointing out each and every flaw ."
      "PROVIDE ONLY A CRTIQUE , NOT A SUGGESTED ANSWER."
      "Then, rate the answer on  a scale of 0 - 100."
      "The response should be in the following format:\n"
      "Critique:<detailed critique>\n"
      "Rating:<rating>"
  )
  rating_response=chat_completion_request(prompt)
  try:
    match=re.search(r'Rating:\s*(\d+)',rating_response)
    if match:
      rating=int(match.group(1))
      if rating>95:
        rating = 95
      rating = float(rating)/100
    else:
        raise ValueError("Rating not found")
  except Exception as e:
    print(f"Error extracting rating:{e}")
    print(f"Rating response was : {rating_response}")
    rating - 0.0


  return rating
```

```python
def get_answer_directly_from_llm(question):
  prompt=(
      f"Question:{question}\n"
      "Please provide  the answer with detailed reasoning. Follow  this format thoroughly : \n"
      "Reasoning Process : <step-by-step reasoning process>\n"
      "Verification:<verification of the facts >\n"
      "Final Answer:<improved and verified answer>\n"
  )

  llm_response=chat_completion_request(prompt)

  #Extract the final answer
  try:
    match=re.search(r'Final Answer:\s*(.*)',llm_response,re.DOTALL)
    final_answer=match.group(1).strip() if match else None
  except Exception as e:
    final_answer=None

  return llm_response,final_answer
```

```python
import math
import random
import numpy as np
```

```python
max_children=3
```

```python
class Node:
  def __init__(self,question,answer,parent=None):
    self.question=question
    self.answer=answer
    self.children=[]
    self.parent=parent
    self.visits=0
    self.value=0.0

  def is_fully_expanded(self):
    return len(self.children)>=max_children

  def best_child(self,exploration_weight=1.41):
    choices_weight=[]
    for child in self.children:
      if child.visits == 0:
        weight=float('inf') # prioritize unexplored nodes
      else:
        weight=(child.value/child.visits)+exploration_weight*math.sqrt((2*math.log(self.visits))/child.visits)
      choices_weight.append((child,weight))
    best_tuple = max(choices_weight, key=lambda x: x[1])
    return best_tuple[0]

  def most_visited_child(self):
    return max(self.children,key=lambda child:child.visits)

  def add_child(self,child):
    self.children.append(child)
```

```python
class MCTS:
  def __init__(self,question,seed_answers,iterations):
    self.seed_answers=seed_answers
    self.iterations=iterations
    self.question=question
    self.root=Node(question,random.choice(seed_answers))

  def search(self):
    for i in range(self.iterations):
      print(f"\nIteration : {i+1}/{self.iterations}")
      node=self.select(self.root)
      print(f"Selected Node : {node.answer}")
      if not node.is_fully_expanded():
        node=self.expand(node)
        print(f"Expanded Node : {node.answer}")
      reward=self.simulate(node)
      print(f"\nSimulated Reward:{reward}")
      self.backpropagate(node,reward)
      print(f"Visits to most visited child:{self.root.most_visited_child().visits}")

    return self.root.most_visited_child().answer

  def select(self,node):
    while node.is_fully_expanded()  and node.children :
      node=node.best_child()
    return node

  def expand(self,node):
    for j in range(max_children-len(node.children)):
      child_node=Node(self.question,node.answer,parent=node) # start with the same answer
      node.add_child(child_node)

      critique= get_critique(self.question,child_node.answer)

      print(f"\n--Critique {j} --\n {critique}")

      improved_answer=improve_answer(self.question,child_node.answer,critique)
      print(f"\n--Improved Answer {j} --\n {improved_answer}")

      child_node.answer=improved_answer # update the child node's answer with the improved answer

    return random.choice(node.children)

  def simulate(self,node):

    rating=rate_answer(self.question,node.answer)

    return rating

  def backpropagate(self,node,reward):
    while node is not None:
      node.visits+=1
      node.value+=reward
      node=node.parent
```

```python
question='What is the captial of France?'
mcts=MCTS(question,seed_answers,iterations=2)
best_answer=mcts.search()
print(f"\nBest Answer:{best_answer}")
```

```python
import re

def extract_boxed_answer(answer):
    """
    Extracts the content within the last \boxed{} in the answer,
    handling nested braces.
    """
    pattern = re.compile(r'\\boxed{((?:[^{}]|\{(?:[^{}]|\{[^{}]*\})*\})*)}')
    matches = pattern.findall(answer)

    if matches:
        return matches[-1]
    return None
```

```python
def get_MATH_QA(row_number=None):
  dataset = load_dataset("DigitalLearningGmbH/MATH-lighteval","default",split='test[:100]')
  df=pd.DataFrame(dataset)
  if row_number is None:
    row_number=int(input("Please enter the row number(0-99) you want to pick:"))

  if row_number<0 or row_number>=len(df):
    raise ValueError("Invalid row number")

  selected_row=df.iloc[row_number]

  question=selected_row['problem']
  full_answer=selected_row['solution']
  short_answer=extract_boxed_answer(full_answer)
  return question,full_answer,short_answer
```

```python
question, full_answer, short_answer=get_MATH_QA()

print("**Fetched Question:**",question)

llm_full_answer, llm_short_answer=get_answer_directly_from_llm(question)

print("\n**Direct LLM Query Answer:**",llm_full_answer)
print("\n**Ground Truth Full Answer:**",llm_full_answer)
print("\n**Ground Truth Short Answer:**",llm_short_answer)
```
